### Data loading and preparation

```python
# Would you like some comments for results sir
VERBOSE = 0

# Reading data path
READING_PATH = 'Data/KibotData_Top_200_most_liquid_Stocks/'
TICKERS = get_ticker_names(READING_PATH)

# Params for the model
hyperparameters_grid = {
    "dropout_rate": [0],#[0.1, 0],
    "learning_rate": [0.01],#[0.001, 0.01, 0.1, 1, 10],
    "epochs": [50]
}

walk_forward_params = {
    'number_of_blocks' : 10,
    'train_blocks' : 6,
    'validation_blocks' : 2,
    'test_blocks' : 1
}
```

### Model - Walk Forward

```python
for ticker in TICKERS:

    print(f'Loading data for {ticker}')
    # Load the data per ticker
    X_ask, y_ask, X_bid, y_bid = get_bid_ask(ticker, path = READING_PATH)

    # Make predictions for bid and ask
    print(f'Calculating ASK for {ticker}')
    results_ask = walk_forward_block_optimization(
        X = X_ask,
        y = y_ask,
        walk_forward_param_grid = walk_forward_params,
        param_grid = hyperparameters_grid,
        train_fn = train_lstm_model,
        predict_fn = predict_lstm,
        verbose = VERBOSE
    )

    print(f'Calculating BID for {ticker}')
    results_bid = walk_forward_block_optimization(
        X = X_bid,
        y = y_bid,
        walk_forward_param_grid = walk_forward_params,
        param_grid = hyperparameters_grid,
        train_fn = train_lstm_model,
        predict_fn = predict_lstm,
        verbose = VERBOSE
    )

    # save .csv
    print(f'Saving the results of {ticker}...')
    save_results(results_ask, results_bid, ticker, walk_forward_params)
    print('-'*50)
```
